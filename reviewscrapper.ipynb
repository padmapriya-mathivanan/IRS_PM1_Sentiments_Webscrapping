```
import requests
import bs4
import csv

res = requests.get('https://www.kiasuparents.com/kiasu/forum/viewforum.php?f=76&sid=fad07fec7b4dab8e489e19fffc529d79') 
soup = bs4.BeautifulSoup(res.text,'lxml')
with open('/content/drive/My Drive/Project/ScrappedTitles.csv','w',newline='') as f :
    thewriter = csv.writer(f)
    thewriter.writerow(['name'])
    for i in soup.select('.topictitle'):
        thewriter.writerow([i.text])
```

```
import requests
from bs4 import BeautifulSoup

URL = 'https://www.kiasuparents.com/kiasu/forum/viewtopic.php?f=76&t=1508'
page = requests.get(URL)
soup = BeautifulSoup(page.content, 'html.parser')
results = soup.find(id='page-body')
title = results.find('h2', class_='topic-title')
print(title)
with open('/content/drive/My Drive/Project/Scrapped_data.csv','a',newline='') as f :
    thewriter = csv.writer(f)
    result=title.text
    for i in soup.select('.content'):
        thewriter.writerow([result,i.text])
```
